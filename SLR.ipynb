```python
m,c=np.random.random() , np.random.random()  
```

```python
def learn(m,c,x,y,epoch):
    for i in range(epoch):
        error = y - (m*x+c)
        x_error = (y-(m*x+c))*x
        error = np.sum(error)/len(x)
        x_error = np.sum(x_error)/len(x)
        learning_rate = 0.01
        delta_m=learning_rate*x_error
        delta_c=learning_rate*error
        m = m+delta_m
        c = c+delta_c
    return (m,c)
```

```python
(m,c)=learn(m,c,x,y,25000)
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
y=m*x+c
plt.plot(x,y)
```
